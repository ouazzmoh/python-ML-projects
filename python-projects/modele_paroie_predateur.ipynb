## 1. Introduction

On souhaite décrire l'évolution de deux populations en interaction: une population de proies
(par exemple des lièvres)
et une population de prédateurs (par exemple des lynx).  On note respectivement $x(t)$ et $y(t)$ les effectifs des
proies et prédateurs 
(normalisés par des valeurs de référence) à l'instant $t$.
On peut modéliser l'évolution de ces populations
à l'aide d'équations différentielles de la forme:
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) \, f(x(t),y(t),t) \\
y'(t) &=& y(t) \, g(x(t),y(t),t)
\end{array}
\right.
\end{equation}
où $f,g$ sont des fonctions $C^1$ qui décrivent les taux de croissance de chaque population.
La fonction $y\mapsto f(x,y,t)$ est décroissante (terme de prédation) et
$x\mapsto g(x,y,t)$ est croissante.
Par ailleurs, on suppose les fonctions 
$x\mapsto f(x,y,t)$ et $y\mapsto g(x,y,t)$ décroissantes
si on veut prendre en compte une quantité limitée de nourriture.
La dépendance de $f,g$ par rapport au temps permet
de tenir compte des variations saisonnières
de la quantité de nourriture et des taux de natalité.

## 2. Equation de Lotka-Volterra

Il s'agit d'une version très simplifiée du modèle précédent, dans laquelle 
la population de proies dispose de ressources illimitées et 
l'équation différentielle est autonome, i.e. les fonctions $f,g$ ne
dépendent pas de $t$: 
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) ( 1- y(t) ) \\
y'(t) &=& \alpha\, y(t)(x(t)  - 1).
\end{array}
\right.
\end{equation}
Le paramètre $\alpha$ est strictement positif.
Pour les calculs numériques, on fixera $\alpha =1$.

### 2.1. Analyse du modèle

#### Question 1 
Calculer la solution $(x(t),y(t))$ pour la condition initiale
$(x(0),y(0))=(0,y_0)$ avec $y_0 \geq 0$. Quel est son comportement quand $t \rightarrow +\infty$ ?
Interpréter ce résultat.

#### Réponse 1

Le système (2) peut être écrit sous la forme d'un problème de Cauchy:
\begin{equation}
\left\{
\begin{array}{lll}
u'(t)&=& g(t, u) \\
u(0) &=& (0, y_0)
\end{array}
\right.
\end{equation}
Avec $u(t) = (x(t), y(t))$ et $g(t, u) = (x(t)(1-y(t)), \alpha y(t)(x(t) - 1)$

On a $g$ une fonction polynomiale, alors g est de $C^1$ sur l'ouvert $\mathbb{R^2}$.
Alors d'après le theorème de Cauchy-Lipchitz, le problème de Cauchy admet une solution unique $u$ de $C^2$

D'autre part on a $x(0) = 0$
Donc on commence par résoudre
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) z(t) \\
x(0) &=& 0
\end{array}
\right.
\end{equation}
avec $z(t) = 1 - y(t)$

Puisque x est de $C^1$ alors ce problème admet une solution unique x :
$$x(t) = \lambda e^{A(t)}$$
Avec A une primitive de y, et $\lambda \in \mathbb{R}$

Puisque $e^{A(t)} > 0 $ alors en utilisant la condition initiale $x(0) = 0$ on trouve que $\lambda = 0$
Donc $$x'(t) = 0$$

Le système (2) devient alors 

\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& 0 \\
y'(t) &=& -\alpha\, y(t).
\end{array}
\right.
\iff
\left\{
\begin{array}{lll}
x(t)&=& \mu\\
y(t) &=& \omega e^{-\alpha t}
\end{array}
\right.
\end{equation}

En utilisant les Conditions initiales on trouve que:
$$\boxed{
\begin{equation}
\left\{
\begin{array}{lll}
x(t)&=& 0 \\
y(t) &=& y_0 e^{-\alpha t}
\end{array}
\right.
\end{equation}}$$

##### Intérpretation

Sous la condition initiale $(x(0), y(0)) = (0, y_0)$, on trouve que quand $t \to +\infty$ la solution $$(x(t),y(t)) = (0,  e^{-\alpha t}) \to (0,0)$$ 
Cela est cohérent car si à $t = 0$ il n'y a pas de proies, les prédateurs n'auront pas de quoi se nourrir. Donc leur croissance tendera vers $0$.

### 2.2. Intégration Numérique

#### 2.2.1. Schéma d'Euler explicite
Afin d'approcher numériquement les solutions du modèle de Lotka-Volterra
pour $t \in [0,t_{\mathrm{max}}]$, 
on discrétise l'équation à l'aide du schéma d'Euler explicite:
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}&=& x_k + h\, x_k ( 1- y_k ) \\
y_{k+1} &=& y_k + h\, y_k (x_k  - 1)
\end{array}
\right.
\end{equation}
où $h >0$ est le pas d'intégration en temps et
$(x_k,y_k)$ une approximation de $(x(t_k),y (t_k))$ en $t_k = k\, h$.
On fixe $h=t_{\mathrm{max}}/N$ où $N$ est un entier $\geq 1$.

#### Question 7
Pour $t_{\mathrm{max}}=120$, $(x(0),y(0))=(2,1)$ 
et différentes valeurs du pas de temps $h$, 
tracer les trajectoires 
$(x_k,y_k)_{0 \leq k \leq N} \subset {\left(\mathbb{R}^+ \right)}^2$
ainsi que les graphes de $x_k$, $y_k$ en fonction de $t_k$. 
Qu'observez-vous ? Comparer ces résultats à l'étude théorique 
de la question 6.

##### Réponse 7

```python
tmax = 120
(x0, y0) = (2, 1)

def euler_explicite(xold, yold, h):
    xnew = xold + h*xold*(1-yold)
    ynew = yold + h*yold*(xold-1)
    return xnew, ynew


for N in [5000, 10000, 100000, 1000000]:
    h = tmax/N
    t = np.linspace(0,tmax, N)
    X = np.zeros(N)
    Y = np.zeros(N)
    X[0], Y[0] = x0, y0
    for i in range(1, N):
        X[i], Y[i] = euler_explicite(X[i-1], Y[i-1], h)

    #Plotting
    plt.subplots(figsize = (15, 10))
    plt.subplot(2,2,1)
    plt.plot(X,Y, "g")
    plt.title(f"Y en fonction X : Euler Explicite   h = {tmax/N}")
    plt.plot(X[0],Y[0],"b*")
    plt.annotate("Condition initiale", (X[0] + 0.02, Y[0] + 0.01), color = "b")
    plt.plot(1,1,"r*")
    plt.annotate("Point d'équilibre", (1.02,1.01), color = "r")
    plt.xlim(0,3)
    plt.ylim(0,3)
    plt.grid()
    plt.subplot(2,2,2)
    plt.plot(t, Y, "r", label = "Y(t): Prédateurs")
    plt.plot(t, X, "g", label = "X(t): Proies")
    plt.legend()
    plt.xlim(0,90)
    plt.ylim(0,3)
    plt.grid()
    plt.title(f"X et Y en fonction du temps:Euler Explicite   h = {tmax/N}")
    plt.show()




```

##### Intérpretaton:
On remarque que pour ce schéma le portrait de phase est une spirale, dont le point initial est à l'intérieur ce qui implique l'irréversibilité puisque parcourir une spirale dans un sens n'est pas identique à la parcourir dans l'autre sens. D'autre part la valeur maximale augmente au cours du temps , contrairement à l'étude théorique qui reste la même .Ainsi , les solutions ne sont plus périodiques mais pseudo-périodiques.

De plus, il faut diminuer largement la pas de discrétisation (donc augmenter $N$) pour arriver à une courbe qui est parfaitement périodique. Donc le schéma explicite est trés coûteux.

#### Réponse 8

$x_{k+1} - h x_{k+1}(1- y_{k+1}) = x_{k} $  => $ \boxed {x_{k+1} =  \frac{x_{k}}{1 - h(1- y_{k+1})}}$

#### Question 9
Reprendre la question 7 pour le schéma d'Euler implicite.

#### Réponse 9

```python
tmax = 120
(x0, y0) = (2, 1)

def euler_implicite(xold, yold, h):
    delta = (1 - h*yold - h*xold - h**2)**2 - 4*(h + h**2)*(h - 1)*yold
    b = 1 - h*yold - h*xold - h**2
    a = h + h**2
    ynew = (-b + np.sqrt(delta))/(2*a)
    xnew = xold/(1 - h*(1-ynew))
    return xnew, ynew

for N in [5000, 10000, 100000, 1000000]:
    h = tmax/N
    t = np.linspace(0,tmax, N)
    X = np.zeros(N)
    Y = np.zeros(N)
    X[0], Y[0] = x0, y0
    for i in range(1, N):
        X[i], Y[i] = euler_implicite(X[i-1], Y[i-1], h)

    #Plotting
    plt.subplots(figsize = (15, 10))
    plt.subplot(2,2,1)
    plt.plot(X,Y, "g")
    plt.title(f"Y en fonction X : Euler Implicite   h = {tmax/N}")
    plt.plot(X[0],Y[0],"b*")
    plt.annotate("Condition initiale", (X[0] + 0.02, Y[0] + 0.01), color = "b")
    plt.plot(1,1,"r*")
    plt.annotate("Point d'équilibre", (1.02,1.01), color = "r")
    plt.xlim(0,2.5)
    plt.ylim(0,2.5)
    plt.grid()
    plt.subplot(2,2,2)
    plt.plot(t, Y, "r", label = "Y(t): Prédateurs")
    plt.plot(t, X, "g", label = "X(t): Proies")
    plt.legend()
    plt.xlim(0,90)
    plt.ylim(0,3)
    plt.grid()
    plt.title(f"X et Y en fonction du temps:Euler Implicite  h = {tmax/N}")
    plt.show()
```

##### Intérpretation
Comme pour le schéma explicite, on remarque que le portrait de phase est une spirale, dont le point initial est à l'éxtérieur (contrairement au cas explicite) ce qui implique l'irréversibilité . D'autre part la valeur maximale diminue  au cours du temps , contrairement à l'étude théorique qui reste la même et le cas explicite qui augmente. Ainsi , les solutions ne sont plus périodiques mais pseudo-périodiques.

Le schéma implicite donne des résultats assez similaires au schéma explicite pour le choix des valeurs de h.

#### 2.2.3. Schéma d'Euler semi-implicite
On considère un schéma d'Euler semi-implicite

#### Question 11
Pour $t_{\mathrm{max}}=120$ et différentes valeurs du pas de temps $h$, 
tracer les trajectoires 
$(x_k,y_k)_{0 \leq k \leq N} $

```python
tmax = 120
(x0, y0) = (2, 1)

def euler_semiImplicite(xold, yold, h):
    xnew = xold*(1 + h*((1-h*xold + h - yold)/(1-h*(xold-1))))
    ynew = yold/(1-h*(xold-1))
    return xnew, ynew

#Plotting
for N in [300, 500, 1000, 10000]:
    h = tmax/N
    t = np.linspace(0,tmax, N)
    X = np.zeros(N)
    Y = np.zeros(N)
    X[0], Y[0] = x0, y0
    for i in range(1, N):
        X[i], Y[i] = euler_semiImplicite(X[i-1], Y[i-1], h)

    #Plotting
    plt.subplots(figsize = (15, 10))
    plt.subplot(2,2,1)
    plt.plot(X,Y, "g")
    plt.title(f"Y en fonction X : Euler Semi-implicite   h = {tmax/N}")
    plt.plot(X[0],Y[0],"b*")
    plt.annotate("Condition initiale", (X[0] + 0.02, Y[0] + 0.01), color = "b")
    plt.plot(1,1,"r*")
    plt.annotate("Point d'équilibre", (1.02,1.01), color = "r")
    plt.xlim(0,3)
    plt.ylim(0,3)
    plt.grid()
    plt.subplot(2,2,2)
    plt.plot(t, Y, "r", label = "Y(t): Prédateurs")
    plt.plot(t, X, "g", label = "X(t): Proies")
    plt.legend()
    plt.xlim(0,90)
    plt.ylim(0,3)
    plt.grid()
    plt.title(f"X et Y en fonction du temps:Euler Semi-implicite   h = {tmax/N}")
    plt.show()
```

##### Intérpretation :
Le schéma Semi-implicite permet d'atteindre des solutions périodiques avec une grande éfficacité des 2 autres schémas d'Euler . Contrairement au deux premières cas , qui montrent une sensibilité de la variabilité de la valeur maximale au cours du temps .

Contrairement au cas des schémas précedents, le semi-implicite ne nécessite pas un pas de décritisation faible pour arriver à une allure périodique. D'où ce schéma est plus efficace que les autres et moins coûteux .

#### Question 13
On fixe $t_{\mathrm{max}}=120$ et $h=10^{-3}$.
Pour différentes valeurs du paramètre $\lambda$,
tracer les graphes de $x_k$, $y_k$ en fonction de $t_k$
pour différentes conditions initiales $(x(0),y(0))$. 
Qu'observez-vous ? 
Interpréter le comportement des solutions lorsque
$t\rightarrow +\infty$
en terme d'évolution des populations, ainsi que l'effet du paramètre $\lambda$.

#### Réponse 13

```python
tmax = 120
def euler_croissanceLimite(xold, yold, cold, h):
    xnew = xold*(1 + h*(1 - yold/(1 - h*(xold- 1)) - cold*xold))
    ynew = yold/(1-h*(xold-1))
    return xnew, ynew


for param in [(2,1,0.0001), (2,2, 0.1), (2, 1, 1), (1, 2, 2), (1, 2, 5), (2.4, 0.6, 10)]:
    (x0, y0, cte) = (param[0], param[1], param[2])
    h = 1e-3
    N = int(tmax/h)
    t = np.linspace(0,tmax, N)
    C = np.zeros(N)
    for i in range(N):
        C[i] = cte * (1 + (1/2)*np.cos(t[i]*np.pi/6))



    X = np.zeros(N)
    Y = np.zeros(N)
    X[0], Y[0] = x0, y0
    for i in range(1, N):
        X[i], Y[i] = euler_croissanceLimite(X[i-1], Y[i-1],C[i-1], h)

    #Plotting
    plt.subplots(figsize = (15, 10))
    plt.subplot(2,2,1)
    plt.plot(X,Y, "g")
    plt.title(f"Y(X): Croissance Limitée avec lambda = {cte} C.I = {(X[0], Y[0])}")
    plt.xlim(0,2.5)
    plt.ylim(0,2.5)
    plt.plot(X[0],Y[0],"b*")
    plt.annotate("Condition initiale", (X[0] + 0.02, Y[0] + 0.01), color = "b")
    plt.grid()
    plt.subplot(2,2,2)
    plt.plot(t, Y, "r", label = "Y(t): Prédateurs")
    plt.plot(t, X, "g", label = "X(t): Proies")
    plt.legend()
    plt.xlim(0,100)
    plt.ylim(0,3)
    plt.title(f"X(t) et Y(t): Croissance Limitée avec lambda = {cte} C.I = {(X[0], Y[0])}")
    plt.grid()
    plt.show()
```

##### Interprétation
$\lambda$ Étant le paramètre qui tient compte de la nouriturre des paroies, on remarque qu'en augmentant la valeur de $\lambda$, la croissance l'effectif des proies atteint un maximum plus faible lors de son oscillation.
Cela impacte l'effectif des prédateurs qui trouvent moins de proies pour se nourrir.

Quand $t \to +\infty$ $x$ et $y$ convergent et se stabilisent dans un état qui dépend des conditions initiales et du $\lambda$.

Si il n'y a pas beacoup de proies, l'effectif des prédateurs tend vers $0$. Ce qui est encore cohérent.
